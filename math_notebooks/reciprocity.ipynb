```python
# all data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import statsmodels as sm
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
X = np.random.randn(500000)
X2 = np.random.randn(500000)
Y = X/3 + X2/3 + np.random.randn(500000)/3

# plot X and Y
plt.scatter(X, Y)
plt.show()


def plot_line(X, Y):
    # calculate the slope of the line of best fit
    slope = np.cov(X, Y)[0, 1] / np.var(X)
    correlation = np.corrcoef(X, Y)[0, 1]**2

    # calculate the intercept of the line of best fit
    intercept = Y.mean() - slope * X.mean()

    # plot the line of best fit
    plt.plot(X, slope * X + intercept, color='red')
    plt.scatter(X, Y)
    print(f'y = {slope/correlation:.2f}x + {intercept/correlation:.2f}')
    plt.legend([f'y = {slope:.2f}x + {intercept:.2f}', 'Data'])
    plt.show()


plot_line(X, Y)

# switch X and Y
plot_line(Y, X)
```
